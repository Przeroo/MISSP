```python
# Basics of Pulp

import pulp  # import biblioteki PuLP do optymalizacji liniowej

# Creating a problem
prob = pulp.LpProblem("Simple_Problem", pulp.LpMaximize)  # tworzenie problemu, chcemy maksymalizowac funkcje celu

# Variables
x = pulp.LpVariable('x', lowBound=0, upBound=4)  # zmienna x z granicami 0-4
y = pulp.LpVariable('y', lowBound=0, upBound=3)  # zmienna y z granicami 0-3

# Objective function
prob += x + y, "Z"  # funkcja celu: maksymalizacja sumy x i y

# Constraints
prob += x - y == 0  # ograniczenie: x i y musza byc takie same
print(prob)  # wyswietlenie problemu
```

```python
# Solving the problem
prob.solve()

# Printing the results
print("x =", x.value())
print("y =", y.value())
```

```python
import pulp

# Define the problem
prob = pulp.LpProblem("The Blending Problem", pulp.LpMinimize)
```

**Reasoning**:
Define the decision variables for each ingredient as specified in the blending problem tutorial.



```python
# Define the decision variables
# The quantities of each ingredient to use
chicken = pulp.LpVariable("Chicken", lowBound=0, cat="Continuous")
beef = pulp.LpVariable("Beef", lowBound=0, cat="Continuous")
mutton = pulp.LpVariable("Mutton", lowBound=0, cat="Continuous")
rice = pulp.LpVariable("Rice", lowBound=0, cat="Continuous")
wheat_bran = pulp.LpVariable("Wheat_Bran", lowBound=0, cat="Continuous")
gel = pulp.LpVariable("Gel", lowBound=0, cat="Continuous")
sardines = pulp.LpVariable("Sardines", lowBound=0, cat="Continuous") # Added new variable
```

**Reasoning**:
Define the objective function, which is to minimize the total cost of the ingredients, and add it to the problem.



```python
# Define the objective function
# Minimize the total cost
prob += 0.013 * chicken + 0.008 * beef + 0.010 * mutton + 0.002 * rice + 0.005 * wheat_bran + 0.001 * gel + 0.020 * sardines, "Total Cost"
```

**Reasoning**:
Add the constraints to the problem, including the total weight constraint, and the percentage constraints for each ingredient. Also, add the new constraint for Vitamin B content.



```python
# Add the constraints

# Total weight constraint
prob += chicken + beef + mutton + rice + wheat_bran + gel + sardines == 100, "TotalWeight"

# Percentage constraints for each ingredient
prob += chicken <= 0.10 * (chicken + beef + mutton + rice + wheat_bran + gel + sardines), "ChickenMax"
prob += beef <= 0.20 * (chicken + beef + mutton + rice + wheat_bran + gel + sardines), "BeefMax"
prob += mutton <= 0.20 * (chicken + beef + mutton + rice + wheat_bran + gel + sardines), "MuttonMax"
prob += rice >= 0.40 * (chicken + beef + mutton + rice + wheat_bran + gel + sardines), "RiceMin"
prob += rice <= 0.80 * (chicken + beef + mutton + rice + wheat_bran + gel + sardines), "RiceMax"
prob += wheat_bran >= 0.06 * (chicken + beef + mutton + rice + wheat_bran + gel + sardines), "WheatBranMin"
prob += wheat_bran <= 0.10 * (chicken + beef + mutton + rice + wheat_bran + gel + sardines), "WheatBranMax"
prob += gel >= 0.04 * (chicken + beef + mutton + rice + wheat_bran + gel + sardines), "GelMin"
prob += gel <= 0.08 * (chicken + beef + mutton + rice + wheat_bran + gel + sardines), "GelMax"
prob += sardines >= 0.01 * (chicken + beef + mutton + rice + wheat_bran + gel + sardines), "SardinesMin"
prob += sardines <= 0.01 * (chicken + beef + mutton + rice + wheat_bran + gel + sardines), "SardinesMax"


# Vitamin B constraint (minimum 0.8 g daily per 100g of food, since we are working with 100g total)
prob += (0.3 * chicken + 2.6 * beef + 2.6 * mutton + 0 * rice + 0 * wheat_bran + 0 * gel + 8.9 * sardines) >= 0.8, "VitaminBMin"
```

## Implement the blending problem (compact version)

### Subtask:
After successfully implementing and modifying the step-by-step version, implement the more compact version of the blending problem solution from the PuLP documentation.


**Reasoning**:
Implement the more compact version of the blending problem solution from the PuLP documentation by defining the problem, variables, objective function, and all constraints in a single code block.



```python
# Create a new PuLP problem instance
prob_compact = pulp.LpProblem("Compact_Blending_Problem", pulp.LpMinimize)

# Define dictionaries for costs, protein, fat, fibre, and vitamin B content
ingredients = ["chicken", "beef", "mutton", "rice", "wheat bran", "gel", "sardines"]

cost = {"chicken": 0.013, "beef": 0.008, "mutton": 0.010, "rice": 0.002, "wheat bran": 0.005, "gel": 0.001, "sardines": 0.020}
protein = {"chicken": 0.10, "beef": 0.20, "mutton": 0.15, "rice": 0.00, "wheat bran": 0.04, "gel": 0.00, "sardines": 0.25}
fat = {"chicken": 0.08, "beef": 0.11, "mutton": 0.10, "rice": 0.01, "wheat bran": 0.01, "gel": 0.00, "sardines": 0.07}
fibre = {"chicken": 0.00, "beef": 0.00, "mutton": 0.00, "rice": 0.10, "wheat bran": 0.15, "gel": 0.00, "sardines": 0.00}
vitamin_b = {"chicken": 0.3, "beef": 2.6, "mutton": 2.6, "rice": 0.0, "wheat bran": 0.0, "gel": 0.0, "sardines": 8.9}


# Create a dictionary of decision variables
ingredient_vars = pulp.LpVariable.dicts("Ingredient", ingredients, lowBound=0, cat="Continuous")

# Define the objective function
prob_compact += (
    pulp.lpSum([cost[i] * ingredient_vars[i] for i in ingredients]),
    "Total Cost",
)

# Add the constraints

# Total weight constraint
prob_compact += pulp.lpSum([ingredient_vars[i] for i in ingredients]) == 100, "TotalWeight"

# Percentage constraints for each ingredient
prob_compact += ingredient_vars["chicken"] <= 0.10 * 100, "ChickenMax"
prob_compact += ingredient_vars["beef"] <= 0.20 * 100, "BeefMax"
prob_compact += ingredient_vars["mutton"] <= 0.20 * 100, "MuttonMax"
prob_compact += ingredient_vars["rice"] >= 0.40 * 100, "RiceMin"
prob_compact += ingredient_vars["rice"] <= 0.80 * 100, "RiceMax"
prob_compact += ingredient_vars["wheat bran"] >= 0.06 * 100, "WheatBranMin"
prob_compact += ingredient_vars["wheat bran"] <= 0.10 * 100, "WheatBranMax"
prob_compact += ingredient_vars["gel"] >= 0.04 * 100, "GelMin"
prob_compact += ingredient_vars["gel"] <= 0.08 * 100, "GelMax"
prob_compact += ingredient_vars["sardines"] >= 0.01 * 100, "SardinesMin"
prob_compact += ingredient_vars["sardines"] <= 0.01 * 100, "SardinesMax"

# Nutrient constraints
prob_compact += (
    pulp.lpSum([protein[i] * ingredient_vars[i] for i in ingredients]) >= 8.0,
    "ProteinMin",
)
prob_compact += (
    pulp.lpSum([fat[i] * ingredient_vars[i] for i in ingredients]) >= 6.0,
    "FatMin",
)
prob_compact += (
    pulp.lpSum([fibre[i] * ingredient_vars[i] for i in ingredients]) >= 2.0,
    "FibreMin",
)

# Vitamin B constraint
prob_compact += (
    pulp.lpSum([vitamin_b[i] * ingredient_vars[i] for i in ingredients]) >= 0.8,
    "VitaminBMin",
)

print(prob_compact)
```

**Reasoning**:
Solve the compact blending problem and print the status, optimal cost, and the optimal values for each ingredient.



```python
# Solve the problem
prob_compact.solve()

# Print the results
print("Status:", pulp.LpStatus[prob_compact.status])
print("Optimal Cost:", pulp.value(prob_compact.objective))
for v in prob_compact.variables():
    print(v.name, "=", v.value())
```
