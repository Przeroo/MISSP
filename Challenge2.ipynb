```python
import pulp
u = [6.0,10.0,4.0,7.0,3.0]
I_nom_mA = [4.0,2.0,2.0,2.0,4.0]
delta_mA = [1.0]*5
I_nom = [x/1000.0 for x in I_nom_mA]
delta = [d/1000.0 for d in delta_mA]

prob = pulp.LpProblem("Minimal_power_bridge", pulp.LpMinimize)
i_vars = [pulp.LpVariable(f"i{i+1}", lowBound=max(1e-9,I_nom[i]-delta[i]), upBound=I_nom[i]+delta[i]) for i in range(5)]
prob += pulp.lpSum([u[i] * i_vars[i] for i in range(5)])
prob += i_vars[0] - i_vars[2] - i_vars[3] == 0
prob += i_vars[1] + i_vars[2] - i_vars[4] == 0
prob.solve(pulp.PULP_CBC_CMD(msg=False))
print([v.varValue for v in i_vars])
```
